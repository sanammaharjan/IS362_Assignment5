### Q3. On February 12th, 2013, which New York area airport had the windiest weather
To answer this question, we are going to import weather.csv.


```python
WEATHER_FILE = FILEPATH + '\\weather.csv'
WEATHER_DATA = pd.read_csv(WEATHER_FILE)
WEATHER_DATA.head(5)
```

```python
# Filtering data based on our requirements
WEATHER_DATA[(WEATHER_DATA['year'] == 2013)
             & (WEATHER_DATA['month'] == 2)
             & (WEATHER_DATA['day']== 12)].sort_values(by='wind_gust', ascending = False).head(5)
```
